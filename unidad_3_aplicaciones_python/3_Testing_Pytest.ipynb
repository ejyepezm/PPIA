### Pytest: El estÃ¡ndar en Python
Es simple. Escribes una funciÃ³n que empieza con `test_` y usas la palabra clave **`assert`**.
*   `assert 2 + 2 == 4` (Pasa âœ…)

### TDD (Test Driven Development)
FilosofÃ­a avanzada:
1.  **Rojo:** Escribes el test PRIMERO (fallarÃ¡ porque la funciÃ³n no existe).
2.  **Verde:** Escribes el cÃ³digo mÃ­nimo para que el test pase.
3.  **Refactor:** Mejoras el cÃ³digo.

```python
# Instalamos pytest (y ipytest para que funcione en Colab)
!pip install pytest ipytest -q

import ipytest
import pytest

# Configuramos ipytest para que corra en el notebook
ipytest.autoconfig()

# --- CÃ“DIGO A PROBAR (La LÃ³gica de Negocio) ---
def clasificar_riesgo(edad, deuda):
    """Clasifica riesgo crediticio: Alto, Medio, Bajo"""
    if deuda > 5000:
        return "Alto"
    if edad < 18:
        return "Rechazado" # Regla de negocio crÃ­tica
    return "Bajo"

# --- LOS TESTS (El GuardiÃ¡n) ---
# Nota: Las funciones de test deben empezar con 'test_'
def test_deuda_alta():
    resultado = clasificar_riesgo(30, 6000)
    assert resultado == "Alto" # Verificamos que sea Alto

def test_menor_edad():
    resultado = clasificar_riesgo(15, 0)
    assert resultado == "Rechazado" # Esta regla es vital

# --- EJECUTAR TESTS ---
print("Ejecutando pruebas...")
ipytest.run()
```
