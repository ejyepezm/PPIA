## 1. Regla #1 de Git: NUNCA en Main

La rama `main` (o `master`) es sagrada. Es la versiÃ³n del cÃ³digo que estÃ¡ funcionando en producciÃ³n.

**El Flujo Profesional (Git Flow):**
1.  **`git pull`**: Actualizas tu copia local.
2.  **`git checkout -b feature/nuevo-modelo`**: Creas una rama nueva para tu trabajo.
3.  **Haces tu magia:** Escribes cÃ³digo, haces commits.
4.  **`git push`**: Subes tu rama a GitHub.
5.  **Pull Request:** Pides a tus compaÃ±eros que revisen tu cÃ³digo.
6.  **Merge:** Una vez aprobado, se fusiona con `main`.

### Â¿QuÃ© es un Conflicto de FusiÃ³n?
Ocurre cuando tÃº y un compaÃ±ero modifican la **misma lÃ­nea del mismo archivo**. Git no sabe cuÃ¡l versiÃ³n elegir y te pide que resuelvas el conflicto manualmente.

```python
# --- SIMULACIÃ“N EN COLAB ---
# Usamos '!' para ejecutar comandos de terminal

# 1. ConfiguraciÃ³n (Necesario una vez)
!git config --global user.name "TuNombre"
!git config --global user.email "tuemail@ejemplo.com"

# 2. Creamos un Repositorio de Juguete
!git init mi_proyecto_simulado
%cd mi_proyecto_simulado

# 3. Creamos el primer archivo en la rama 'main'
!echo "print('Modelo v1')" > train.py
!git add train.py
!git commit -m "feat: VersiÃ³n inicial del modelo"

print("--- Historial en main ---")
!git log --oneline

# 4. Creamos una rama para una nueva feature
!git checkout -b feature/algoritmo-mejorado

# 5. Hacemos cambios en la rama
!echo "print('Modelo v2 mejorado')" > train.py
!git commit -am "feat: AÃ±adido nuevo algoritmo"

print("\n--- Historial en la rama nueva ---")
!git log --oneline

# 6. Volvemos a main y fusionamos (merge)
!git checkout main
!git merge feature/algoritmo-mejorado

print("\n--- Historial final en main ---")
!git log --oneline
```
